## YouTube ambient music scraping code
### Note: Downloads may take a long time because of song length. QOL improvement: only download videos under a certain duration

```python
from urllib2 import urlopen
import re

# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# ~~~~~~~~~~~~~~~YouTube search code~~~~~~~~~~~~~~~~~
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

def search_for_ambient_music(category):

    import sys
    default_stdout = sys.stdout
    default_stderr = sys.stderr
    reload(sys)
    sys.setdefaultencoding('utf8')

    textToSearch = category + ' ambient music'
    query_string = urllib.quote(textToSearch)
    html_content = urllib2.urlopen("http://www.youtube.com/results?search_query=" + query_string)
    hrefs = re.findall(r'href=\"\/watch\?v=(.{11})', html_content.read().decode())
    search_results = []
    for result in hrefs:
        search_results.append("http://www.youtube.com/watch?v=" + result)

    sys.stdout = default_stdout
    sys.stderr = default_stderr

    count = 0
    for result in search_results:
        count = count + 1
        print("Downloading result " + str(count) + ": " + result)
        download_audio(result, category)
    
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# ~~~~~~~~~~~~~Audio downloading code~~~~~~~~~~~~~~~~
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Need youtube_dl for this. Try 'pip install --upgrade youtube_dl' from command line for Linux.
        
from __future__ import unicode_literals
import youtube_dl


class MyLogger(object):
    def debug(self, msg):
        pass

    def warning(self, msg):
        pass

    def error(self, msg):
        print(msg)


def my_hook(d):
    if d['status'] == 'finished':
        print('Done downloading, now converting ...')


def download_audio(videoURL, category='misc'):
    ydl_opts = {
        'outtmpl': '/home/sbordovsky/Notes/EE 379K/UT-Data-Science-Lab/HW8/' + category + '/foo_%(title)s-%(id)s.%(ext)s',
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'logger': MyLogger(),
        'progress_hooks': [my_hook],
    }
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([videoURL])
```
